```python
# Tarea #1: Limpieza y análisis de datos
# Estudiante: Wilmer Estuardo Vasquez Raxon
# Carné: 201800678

import pandas as pd
import numpy as np

# Cargar el dataset
df = pd.read_csv('dataset_sucio.csv')
print("¡Dataset cargado exitosamente!")
```

```python
print("--- ESTADO ORIGINAL DEL DATAFRAME ---")
display(df.head(10))

print("\n--- TABLA PIVOTE (DATOS SUCIOS) ---")
# Intentamos ver el promedio de gasto por ciudad antes de limpiar
try:
    pivot_sucio = df.pivot_table(index='ciudad', values='gasto_q', aggfunc='mean')
    display(pivot_sucio)
except Exception as e:
    print("Error al crear tabla: Los datos están muy sucios (posiblemente gasto_q no es numérico)")
```

```python
# 1. Eliminación de duplicados
df = df.drop_duplicates()

# 2. Limpieza de strings (Nombre, Ciudad, Categoría)
# Quitamos espacios en blanco extra y estandarizamos a mayúsculas iniciales
cols_texto = ['nombre', 'ciudad', 'categoria']
for col in cols_texto:
    df[col] = df[col].astype(str).str.strip().str.title()

# 3. Corrección de la columna Gasto_Q (De texto a número)
# Reemplazamos la coma por punto para que Python lo reconozca como decimal
df['gasto_q'] = df['gasto_q'].astype(str).str.replace(',', '.')
df['gasto_q'] = pd.to_numeric(df['gasto_q'], errors='coerce')

# 4. Tratamiento de valores faltantes (NaN)
# Llenamos gastos vacíos con el promedio de la columna
df['gasto_q'] = df['gasto_q'].fillna(df['gasto_q'].mean())

# 5. Estandarización de Género
# Pasamos todo a mayúsculas y limpiamos espacios para que 'f' y ' F ' sean 'F'
df['genero'] = df['genero'].astype(str).str.strip().str.upper()
# Opcional: Corregir errores si hay valores extraños
df['genero'] = df['genero'].replace({'MALE': 'M', 'FEMALE': 'F', ' ': 'N/A'})

# 6. Estandarización de Fechas
# Convertimos a formato datetime de pandas
df['fecha_registro'] = pd.to_datetime(df['fecha_registro'], errors='coerce')

print("Limpieza completada.")
```

```python
print("--- DATOS LIMPIOS (DEPURADOS) ---")
display(df.head(10))

print("\n--- ANÁLISIS: GASTO PROMEDIO POR CIUDAD Y GÉNERO ---")
# Esta es la tabla pivote que demuestra que los datos ya están estandarizados
pivot_limpio = df.pivot_table(index='ciudad', columns='genero', values='gasto_q', aggfunc='mean')
display(pivot_limpio)

# Guardar el resultado limpio para descargar
df.to_csv('dataset_limpio.csv', index=False)
```
